```python
import numpy as np

# Tạo mảng 3x3 ngẫu nhiên [0, 20]
arr = np.random.randint(0, 21, size=(3, 3))
print("Mảng:\n", arr)

# Bài 1: Kiểm tra TẤT CẢ phần tử khác 0 [cite: 572, 573]
print("Tất cả phần tử khác 0?", np.all(arr != 0))

# Bài 2: Kiểm tra CÓ ÍT NHẤT một phần tử khác 0 [cite: 573, 574]
print("Có ít nhất một phần tử khác 0?", np.any(arr != 0))
```

```python
# Câu 22: Tạo DataFrame [cite: 707-711]
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(exam_data, index=labels)

# Câu 23: 3 dòng đầu [cite: 712]
print("3 dòng đầu:\n", df.head(3))

# Câu 24: Chọn cột name và score [cite: 713]
print("Cột name & score:\n", df[['name', 'score']])

# Câu 25: Attempts > 2 [cite: 714]
print("Attempts > 2:\n", df[df['attempts'] > 2])

# Câu 26: Số dòng, số cột [cite: 715]
print("Shape:", df.shape)

# Câu 27: Score là NaN [cite: 716]
print("Score NaN:\n", df[df['score'].isnull()])

# Câu 28: Score trong khoảng (15, 20) [cite: 717]
print("Score (15, 20):\n", df[df['score'].between(15, 20)])

# Câu 29: Attempts > 2 và Score (15, 20) [cite: 718]
print("Kết hợp 2 điều kiện:\n", df[(df['attempts'] > 2) & (df['score'].between(15, 20))])

# Câu 30: Thay đổi giá trị dòng 'd', score = 19 [cite: 719]
df.loc['d', 'score'] = 19
print("Dòng d sau khi sửa:\n", df.loc['d'])
```
