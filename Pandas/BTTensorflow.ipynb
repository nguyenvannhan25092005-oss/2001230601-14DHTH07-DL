```python
#Bài 13&14
class BasicMathModel(tf.Module):
    @tf.function
    def compute(self, x, y):
        return x + y, x - y, x * y

model = BasicMathModel()
print("B13&14:")
x = float(input("Nhập số thứ nhất (x): "))
y = float(input("Nhập số thứ hai (y): "))

tong, hieu, tich = model.compute(tf.constant(x), tf.constant(y))

print(f"Tổng: {tong.numpy()}")
print(f"Hiệu: {hieu.numpy()}")
print(f"Tích: {tich.numpy()}")
```

```python
#Bài 15
class PerimeterModel(tf.Module):
    @tf.function
    def __call__(self, a, b):
        return (a + b) * 2.0

model = PerimeterModel()
print("B15:")
a = float(input("Nhập chiều dài (a): "))
b = float(input("Nhập chiều rộng (b): "))

p = model(tf.constant(a), tf.constant(b))
print(f"Chu vi hình chữ nhật là: {p.numpy()}")
```

```python
#Bài 16
class BoxAreaModel(tf.Module):
    @tf.function
    def __call__(self, a, b, c):
        s_xq = (a + b) * c * 2.0
        s_tp = (a * b) * 2.0 + s_xq
        return s_tp

model = BoxAreaModel()
print("B16:")
a = float(input("Nhập chiều dài đáy (a): "))
b = float(input("Nhập chiều rộng đáy (b): "))
c = float(input("Nhập chiều cao (c): "))

s = model(tf.constant(a), tf.constant(b), tf.constant(c))
print(f"Diện tích toàn phần hình hộp là: {s.numpy()}")
```

```python
#Bài 17&18
class TrapezoidProof(tf.Module):
    @tf.function
    def is_equal(self, side_a, side_b):
        return tf.equal(side_a, side_b)

model = TrapezoidProof()
print("B17&18:")
val_1 = float(input("Nhập độ dài cạnh thứ nhất: "))
val_2 = float(input("Nhập độ dài cạnh thứ hai: "))

check = model.is_equal(tf.constant(val_1), tf.constant(val_2))

if check.numpy():
    print("Hai cạnh bằng nhau, thỏa mãn điều kiện chứng minh.")
else:
    print("Hai cạnh không bằng nhau.")
```

```python
#Bài 19
class TriangleMidlineProof(tf.Module):
    @tf.function
    def check_n_midpoint(self, am, mb, parallel_status):
        # parallel_status là True nếu MN // BC
        is_m_midpoint = tf.equal(am, mb)
        return tf.logical_and(is_m_midpoint, parallel_status)

model = TriangleMidlineProof()
print("B19: Chứng minh NA = NC ---")
am = float(input("Nhập độ dài đoạn AM: "))
mb = float(input("Nhập độ dài đoạn MB: "))
parallel_input = input("Đoạn MN có song song với BC không? (c/k): ")

is_parallel = tf.constant(True if parallel_input.lower() == 'c' else False)
result = model.check_n_midpoint(tf.constant(am), tf.constant(mb), is_parallel)

if result.numpy():
    print("NA = NC (N là trung điểm của AC).")
else:
    print("Không đủ điều kiện chứng minh NA = NC.")
```
